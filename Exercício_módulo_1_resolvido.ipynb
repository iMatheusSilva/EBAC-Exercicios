## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
# (A)

qtd_vendas = 3
ticket_medio = 350.52

valor_total_vendas = (qtd_vendas * ticket_medio)

print(f"(A) = {valor_total_vendas}")
```

```
# (B)

valor_total_vendas = 834.47
ticket_medio = 119.21

qtd_vendas = (valor_total_vendas // ticket_medio)

print(f"(B) = {qtd_vendas}")
```

```
# (C)

valor_total_vendas = 15378.12
qtd_vendas = 5

ticket_medio = (valor_total_vendas / qtd_vendas)

print(f"(C) = {ticket_medio}")
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
```

```
# Returns a string where a specified value is replaced with a specified value

cancao_alterada = cancao.replace("mundo", "pneu")
print(cancao_alterada)
```

```
#Splits the string at the specified separator, and returns a list

palavras_da_cancao = cancao.split(" ")
print(palavras_da_cancao)
```

```
# Converts the first character of each word to upper case

print(cancao.title())
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
```

```
palavras_noticia = noticia.split(" ")
selic = palavras_noticia[3]
ano = palavras_noticia[-2]
print(f"Selic = {selic}")
print(f"Ano = {ano}")
```
